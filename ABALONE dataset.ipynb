# Data preprocessing

```python
#now to do  target variable covert into binarization
y = df["Rings"].apply(lambda y_value: 1 if y_value>=7 else 0)
y
```

```python
from imblearn.over_sampling import SMOTE
SM = SMOTE()
x , y = SM.fit_resample(x,y)
```

```python
y.value_counts()
```

# Model Building

```python
#import model building necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
```

```python
#train test split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=12)
```

```python
logistic_reg = LogisticRegression()
random_forest = RandomForestClassifier()
decision_tree = DecisionTreeClassifier()
svm = SVC()
```

```python
#using define function
def classify(model,x,y):
#train the model
    model.fit(xtrain,ytrain)
    print('Accuracy:',model.score(xtest,ytest)*100)
#cross validation    
    score = cross_val_score(model,x,y,cv=5)
    print('CV Score:',np.mean(score)*100)
```

```python
model = LogisticRegression()
classify(model,x,y)
```

```python
model = RandomForestClassifier()
classify(model,x,y)
```

```python
model = DecisionTreeClassifier ()
classify(model,x,y)
```

```python
model = SVC ()
classify(model,x,y)
```
